```python
import xml.etree.ElementTree as ET
from math import radians, cos, sin, asin, sqrt
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tree = ET.parse('race1.gpx')
root = tree.getroot()
```

calculate distance between two gps coordinates

```python
def dist(long1, lat1, elevation1, long2, lat2, elevation2):
    o1, a1, o2, a2 = map(radians, [long1, lat1, long2, lat2])
    
    # difference between longitudes and latitudes
    do = o2-o1
    da = a2-a1
    
    #haversine formula
    a = sin(da/2)**2 + cos(a1) * cos(a2) * sin(do/2)**2
    c = 2*asin(sqrt(a))
    r = 6371 + (elevation1 + elevation2)/2
    return c*r
```

zip root[1][1] with itself, but **offset by one** to allow for calculating distances and times between datapoints

```python
def get_info(root):
    for t1, t2 in zip(root[1][1][1:], root[1][1][:-1]):
        # to make it easier to parse as floats
        vals = (t1.get('lon'), t1.get('lat'), t1[0].text,
                   t2.get('lon'), t2.get('lat'), t2[0].text)
        d = dist(*map(float, vals)) # distance in kilometers
        # timesteps are always 1s, to confirm this I was
        # going to assert it, but i cant get the code to work.
    #     print(time.strptime(t1[1].text, '%y-%m-%dT%H:%M:%SZ'))
        d = d*3600 # convert km/s into km/h
        # order is: distance to next, time, power, atemp, hr, cad
        # todo: parse datetime.
        yield (d, t1[1].text, float(t1[2][0].text), *map(lambda x : float(x.text), t1[2][1]))
```
